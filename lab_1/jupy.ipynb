```python
import cv2
import matplotlib.pyplot as plt


cam = cv2.VideoCapture('/home/anthonyp57/VSCode_projects/interakcja/lab_3/man_vid.mp4')

while(True):
    ret, frame = cam.read()

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(frame)
    plt.show()

    # cv2.imshow('frame',frame)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

    break # to exit after 1 frame
```

```python
import 'package:flutter/material.dart';

void main() {
  runApp(MyApp());
}

class MyApp extends StatelessWidget {
  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'Hellow world',
      theme: ThemeData(
        primarySwatch: Colors.blue,
        visualDensity: VisualDensity.adaptivePlatformDensity,
      ),
      home: MyHomePage(title: 'Hello wordl app'),
    );
  }
}

class MyHomePage extends StatefulWidget {
  MyHomePage({Key? key, required this.title}) : super(key: key);
  final String title;
  @override
  _MyHomePageState createState() => _MyHomePageState();
}

class _MyHomePageState extends State<MyHomePage> {
  String _my_string = ""; // This will store the result of addition
  int _a = 0; // First number
  int _b = 0; // Second number

  void _myOnClickFunction() {
    setState(() {
      _my_string = (_a + _b).toString(); // Add numbers and store the result
    });
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(
        title: Text(widget.title),
      ),
      body: Center(
        child: Column(
          mainAxisAlignment: MainAxisAlignment.center,
          children: <Widget>[
            TextField(
              onChanged: (value) => _a = int.tryParse(value) ?? 0, // Get the first number
              decoration: InputDecoration(
                border: OutlineInputBorder(),
                labelText: 'A',
              ),
            ),
            TextField(
              onChanged: (value) => _b = int.tryParse(value) ?? 0, // Get the second number
              decoration: InputDecoration(
                border: OutlineInputBorder(),
                labelText: 'B',
              ),
            ),
            Text(
              '$_my_string', // Display the result
              style: Theme.of(context).textTheme.headline4,
            ),
            OutlineButton(
              child: new Text("Click me"), // Button to perform addition
              onPressed: _myOnClickFunction,
            )
          ],
        ),
      ),
    );
  }
}
```
